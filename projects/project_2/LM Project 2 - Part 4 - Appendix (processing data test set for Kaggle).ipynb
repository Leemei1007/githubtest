### Data cleaning - test set 

```python
datatest = pd.read_csv('./datasets/test.csv')

#to be able to see all columns
pd.set_option('display.max_columns', None)
```

```python
#checking what's inside:

datatest.head()
```

```python
#checking for nulls

datatest.info()
```

```python
datatest['Alley'].fillna(0, inplace = True)

datatest['Alley'] = datatest['Alley'].replace('Pave', 1)
datatest['Alley'] = datatest['Alley'].replace('Grvl', 1)
```

```python
for column in datatest.select_dtypes(exclude=['object']):
    datatest[column] = datatest[column].fillna(value = 0)
```

```python
for column in datatest.select_dtypes(include=['object']):
    datatest[column] = datatest[column].fillna(value = 'None')
```

```python
datatest.info()
```

Categorical data

```python
# adding columns for building age (at the time it was sold) and renovation/remodelled age

datatest['Building age'] = datatest['Yr Sold'] - datatest['Year Built']
datatest['Remodel age'] = datatest['Yr Sold'] - datatest['Year Remod/Add']
```

```python
# dropping the columns identified earlier
# Since the building age and remodelled age are available, I shall remove the years too

drop_columns = ['PID',
                'Street', 
                'Utilities', 
                'Condition 2', 
                'Roof Matl', 
                'Heating',
                'BsmtFin Type 2',
                'Garage Cond', 
                'Misc Feature', 
                'Yr Sold',
                'Year Remod/Add',
                'Year Built', 
                'Garage Yr Blt',
                'Pool QC',
                'Fireplace Qu']

datatest.drop(columns= drop_columns, inplace=True)
```

```python
# lot shape

def lot_shape(x):
    if x == 'IR3':
        return 0
    elif x == 'IR2':
        return 1 
    elif x == 'IR1':
        return 2 
    elif x == 'Reg':
        return 3

datatest['Lot Shape'] = datatest['Lot Shape'].map(lot_shape)
```
